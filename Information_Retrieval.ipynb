```
#Proses Crawling

my_file = open("crawled.txt", "w+")

for doc_id, article in enumerate(source):
    sources = requests.get(article).text
    soup = BeautifulSoup(sources,'lxml')
    #print(soup.prettify())

    for title, content in zip(soup.find_all('h1', class_ = 'read__title'), soup.find_all('div', class_ = 'read__content')):
        
        my_file.write('<DOC>\n')
        my_file.write(f'<ID>{doc_id + 1}</ID>\n')
        my_file.write(f'<TITLE>{title.text}</TITLE>\n')      
        #print(f'<TITLE>{title.text}</TITLE>')

        summary = content.find_all('p')
        clear_sum = ' '.join([element.text for element in summary])
        my_file.write(f'<TEXT>{clear_sum}</TEXT>\n')
        my_file.write('</DOC>\n')

crawled = open("crawled.txt", "r").read()
```

```
#Processing Crawled Document

#Merubah format dokumen menjadi list of dictionary yang dipisah berdasarkan title dan text
def extract_document(document):
    corpus = []
    title = re.compile(r'<TITLE>(.*?)<\/TITLE>', re.DOTALL)
    text = re.compile(r'<TEXT>(.*?)<\/TEXT>', re.DOTALL)
    for doc in re.findall(r'<DOC>(.*?)<\/DOC>', document, re.DOTALL):
        dokumen = {}
        dokumen['TITLE'] = title.search(doc).group(1)
        dokumen['TEXT'] = text.search(doc).group(1)
        corpus.append(dokumen)
    return corpus

#Menghapus enter, tab, angka, dan kode
def cleaning(ekstrak):
    hapus_enter = re.sub(r'\n+', " ", ekstrak)
    hapus_tab = re.sub(r'\t+', " ", hapus_enter)
    hapus_angka = re.sub(r'\d+', " ", hapus_tab)
    hapus_kode = re.sub(r'[^\x00-\x7F]+'," ", hapus_angka)
    hapus = hapus_kode.translate(str.maketrans("","",string.punctuation))
    return hapus

#Merubah tiap karakter menjadi huruf kecil  
def casefolding(ekstrak):
  word = ekstrak.lower()
  return word

#Merubah kata menjadi bentuk token
def tokenisasi(ekstrak):
  token = ekstrak.split()
  return token

pureDocuments = extract_document(crawled)
documents = extract_document(crawled)
for doc in documents:
    doc['TITLE'] = cleaning(doc['TITLE'])
    doc['TEXT'] = cleaning(doc['TEXT'])

    doc['TITLE'] = casefolding(doc['TITLE'])
    doc['TEXT'] = casefolding(doc['TEXT'])

    doc['TITLE'] = tokenisasi(doc['TITLE'])
    doc['TEXT'] = tokenisasi(doc['TEXT'])

print(documents)
```

```
#Method untuk penggabungan kata dengan AND dan OR

# Memilah 2 kumpulan index dokumen, menjadi hanya kumpulan index yang terdapat pada kedua kumpulan sebelumnya (operator AND)
def operatorAND(num1, num2):
    num = num1 + num2
    size = len(num)
    repeated = []
    for i in range(size):
        k = i + 1
        for j in range(k, size):
            if num[i] == num[j] and num[i] not in repeated:
                repeated.append(num[i])
    return repeated

# Memilah 2 kumpulan index dokumen, menjadi gabungan kumpulan index sebelumnya tanpa ada duplikasi index (operator OR)
def operatorOR(num1, num2):
    num = num1 + num2
    size = len(num)
    unique = []
    for i in range(size):
        if num[i] not in unique:
            unique.append(num[i])
    return unique

# Implementasi operator AND
def implementAND(query):
    index = 0
    while index < len(query):
        word = query[index]
        if isinstance(word, str) and word.lower() == "and":
            query[index] = operatorAND(query[index - 1], query[index + 1])
            query.pop(index + 1)
            query.pop(index - 1)
        else:
            index += 1
    return query

# Implementasi operator OR
def implementOR(query):
    index = 0
    while index < len(query):
        word = query[index]
        if isinstance(word, str) and word.lower() == "or":
            query[index] = operatorOR(query[index - 1], query[index + 1])
            query.pop(index + 1)
            query.pop(index - 1)
        else:
            index += 1
    return query
```
